## Statistical inference

Confidence Interval

```python
import numpy as np

np.random.seed(42)
height = list(np.round(np.random.normal(180, 10, 1000)))
```

```python
print(height)
```

```python
import scipy.stats as st
```

```python
ci = st.norm.interval(confidence=.9, loc=np.mean(height), scale=st.sem(height))

print(np.round(ci, 2))
```

```python
height_mean = np.mean(height)
print(np.round(height_mean, 2))
```

Hypothesis testing

```python
t_statistics, p_value = st.ttest_1samp(height, 182)

print(p_value)
```

```python
if p_value < .05:
    print('We can reject the null hypothesis.')
else:
    print('We cannot reject the null hypothesis.')
```
