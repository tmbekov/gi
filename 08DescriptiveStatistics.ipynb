### Categorical Data. Bar chart

```python
import matplotlib.pyplot as plt
```

```python
year = ['First', 'Second', 'Third', 'Fouth', 'Fifth']
frequency = [500, 470, 300, 260, 220]

plt.bar(year, frequency)

plt.xlabel('Course')
plt.ylabel('Number of students')
plt.title('Distribution of students by university courses')

plt.show()
```

### Quantitative data. The growth of men in Russia

```python
import numpy as np
```

```python
np.random.seed(42)
height = list(np.round(np.random.normal(180, 10, 1000)))
```

```python
print(height)
```

```python
bins = 10

plt.hist(height, bins)

plt.xlabel('Height, cm')
plt.ylabel('Number')
plt.title("The distribution of men's height in Russia")

plt.show()
```

Calculate quantitative values

```python
import statistics
```

```python
np.round(statistics.mean(height), 2)
```

```python
statistics.median(height)
```

```python
statistics.mode(height)
```

```python
statistics.stdev(height)
```

Difference between median and average

```python
salary = [75, 86, 90, 83, 100]

print(statistics.mean(salary), statistics.median(salary))
```

```python
salary_new = [75, 86, 90, 83, 200]
print(statistics.mean(salary_new), statistics.median(salary_new))
```
