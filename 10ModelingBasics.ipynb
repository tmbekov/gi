The relationship of variables

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

np.random.seed(42)
height_men = list(np.round(np.random.normal(180, 10, 1000)))
height_women = list(np.round(np.random.normal(160, 10, 1000)))
```

```python
print(height_men)
print(height_women)
```

```python
bins = 15

plt.figure(figsize=(10, 6))

plt.hist(height_men, bins, alpha=0.5, label='Men height')
plt.hist(height_women, bins, alpha=0.5, label='Women height')

plt.legend(loc='upper left', prop={'size': 14})

plt.xlabel('Height, sm')
plt.ylabel('Number')
plt.title('The distribution of the height of men and women in Russia', fontsize=18)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(height_women_new, neck_women)

plt.xlabel('Height, sm', fontsize=16)
plt.ylabel('Neck circumference, sm', fontsize=16)
plt.title('The dependence of height and neck circumference among women in Russia')

plt.show()
```

Correlation of two variables

```python
import scipy.stats as st

np.round(st.pearsonr(height_women_new, neck_women)[0], 2)
```

Building a model. Regression model

```python
X = np.array(height_women_new)
y = np.array(neck_women)

plt.figure(figsize=(10, 6))

plt.scatter(X, y)

slope, intercept = np.polyfit(X, y, 1)

plt.plot(X, X * slope + intercept, 'r')

plt.xlabel('Women height, sm', fontsize=16)
plt.ylabel('Neck circumference', fontsize=16)
plt.title('The dependence of height and neck circumference among women in Russia', fontsize=18)

plt.show()
```

```python
print(np.round(slope, 2), np.round(intercept, 2))
```
